Q.1  Handle Missing Values (Basic Cleaning)

Fill missing values using the following:

Age → Median

Salary → Mean

Hours_Worked_Per_Week → Median

Performance_Score → Mean

Display the dataset after handling missing values.


...>> ✅ Steps:
Fill Age → Median
Fill Salary → Mean
Fill Hours_Worked_Per_Week → Median
Fill Performance_Score → Mean
💻 Python Code:
import pandas as pd

# Load dataset (replace with your file path if needed)
df = pd.read_csv("your_dataset.csv")

# Fill missing values
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
df['Hours_Worked_Per_Week'].fillna(df['Hours_Worked_Per_Week'].median(), inplace=True)
df['Performance_Score'].fillna(df['Performance_Score'].mean(), inplace=True)

# Display cleaned dataset
print(df)
📊 Optional: Check if any missing values remain
print(df.isnull().sum())
💡 Why this approach?
Median is used for Age & Hours_Worked because it is robust to outliers.
Mean is used for Salary & Performance_Score assuming a relatively normal distribution.

Q.7  Build Preprocessing Pipeline

Create a pipeline that:

Applies encoding to categorical columns

Applies scaling to numerical columns

Use:

ColumnTransformer

Pipeline

.....>> To build a clean and reusable preprocessing pipeline, you can combine scikit-learn’s ColumnTransformer and Pipeline. This lets you handle categorical encoding + numerical scaling in one step.

✅ Example Setup

Assume:

Categorical columns → Gender, Department, Work_Mode, Location
Numerical columns → Age, Salary, Projects_Completed, Hours_Worked_Per_Week
✅ Code Implementation
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Define columns
categorical_cols = ['Gender', 'Department', 'Work_Mode', 'Location']
numerical_cols = ['Age', 'Salary', 'Projects_Completed', 'Hours_Worked_Per_Week']

# Create transformers
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
numerical_transformer = StandardScaler()

# Combine using ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_cols),
        ('num', numerical_transformer, numerical_cols)
    ]
)

# Create full pipeline
pipeline = Pipeline(steps=[
    ('preprocessing', preprocessor)
])

# Apply pipeline
processed_data = pipeline.fit_transform(df)

# Convert to DataFrame (optional for display)
encoded_cols = pipeline.named_steps['preprocessing']\
    .named_transformers_['cat']\
    .get_feature_names_out(categorical_cols)

all_columns = list(encoded_cols) + numerical_cols

processed_df = pd.DataFrame(processed_data, columns=all_columns)

# Display result
print(processed_df.head())
🧠 What this pipeline does
Categorical columns → converted using One-Hot Encoding
Numerical columns → scaled using StandardScaler
Everything is handled in one consistent workflow
🚀 Why this is important
Prevents data leakage in ML models
Keeps preprocessing consistent for training & testing
Easily integrates with models like:
from sklearn.linear_model import LinearRegression

model_pipeline = Pipeline(steps=[
    ('preprocessing', preprocessor),
    ('model', LinearRegression())
]).

Q.9 Conceptual Question

Why is scaling important in Python?

....>> 🧠 Why scaling is important

1. Ensures fair contribution of features
If one feature (e.g., Salary = 100000) is much larger than another (Age = 25), models may treat Salary as more important just because of its scale—not because it actually is.

2. Improves model performance
Many algorithms depend on distance calculations, such as:

KNN
K-Means
SVM

Without scaling, distance-based methods become biased toward large-value features.

3. Faster convergence in optimization
Algorithms like Gradient Descent (used in Linear/Logistic Regression, Neural Networks) converge faster when data is scaled, because:

The cost function becomes smoother
Updates are more stable

4. Required for certain algorithms
Some models assume data is centered or normally distributed:

Linear Regression (better stability)
PCA (Principal Component Analysis)
SVM

Scaling (especially Standardization) helps meet these assumptions.

5. Handles unit differences
Features may have different units:

Age → years
Salary → rupees
Distance → kilometers

Scaling brings everything to a comparable range.

⚖️ When scaling matters most
✅ Distance-based models
✅ Gradient-based optimization
❌ Tree-based models (like Decision Trees, Random Forest) usually don’t need scaling
🚀 Simple intuition

Without scaling → one feature “dominates”
With scaling → all features “compete fairly”.

Q.10 Conceptual Question

Why do we convert categorical data into numerical form?

...>> 🧠 Why this conversion is necessary

1. Algorithms require numerical input
Models like Linear Regression, KNN, and SVM perform mathematical operations (distance, dot products, gradients). Text labels can’t be used directly in these calculations.

2. Enables meaningful comparisons
Numerical encoding allows the model to:

Measure similarity/distance
Detect patterns across categories

Without conversion, categories are just strings with no mathematical meaning.

3. Helps models learn patterns
Encoding transforms categories into a format where relationships can be captured. For example:

One-Hot Encoding → treats each category independently
Label Encoding → assigns ordered values (useful only when order exists)

4. Avoids incorrect assumptions
If you directly assign numbers carelessly (e.g., IT=1, HR=2), the model might assume HR > IT, which is meaningless. Proper techniques like One-Hot Encoding prevent this issue.

5. Required for pipeline integration
In real-world ML workflows, preprocessing pipelines (encoding + scaling) must output purely numeric data before feeding it into models.

⚖️ Common Encoding Methods
One-Hot Encoding → Best for nominal data (no order)
Label Encoding → Used when categories have order (e.g., Low < Medium < High)
🚀 Simple intuition

Categorical data = human-readable
Numerical data = machine-readable

So encoding acts like a translator between humans and algorithms.
